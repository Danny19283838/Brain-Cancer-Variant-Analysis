**Extraction of Chromosome Position, Reference Gene, and Alternative Gene from Data**

```python
import re
import pandas as pd

positions = []
references = []
alternatives = []

for hgvs in df['HGVS_5']:
    # Convert the entry to a string to handle non-string values (like NaN)
    hgvs_str = str(hgvs)
    match = re.search(r':g\.(\d+)([A-Z])>([A-Z])', hgvs_str)
    if match:
        positions.append(match.group(1))
        references.append(match.group(2))
        alternatives.append(match.group(3))
    else:
        positions.append(None)
        references.append(None)
        alternatives.append(None)

result_df = pd.DataFrame({
    'pos': positions,
    'ref': references,
    'alt': alternatives
})

display(result_df)
```

```python
# Save the extracted data DataFrame to a CSV file
result_df.to_csv('extracted_variant_data.csv', index=False)

print("Extracted dataset saved as 'extracted_variant_data.csv'")
```

```python
# Filter out rows where 'pos' is None
filtered_result_df = result_df.dropna(subset=['pos']).copy()

# Convert 'pos' to integer to prevent scientific notation in output
filtered_result_df['pos'] = filtered_result_df['pos'].astype(int)

# Save the extracted data DataFrame to a CSV file
filtered_result_df.to_csv('extracted_variant_data.csv', index=False)

print("Extracted dataset saved as 'extracted_variant_data.csv'")
```
